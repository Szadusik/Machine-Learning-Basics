# Imports

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

# Utilities & data generation

```python
def generate_points(side,dimension,amount): #Generate some amount of points with given dimension included inside hipercube
    res = np.array([])
    for i in range(amount):
        points = np.array([])
        for j in range(dimension):
            points = np.append(points,random.uniform(-side,side))
        res = np.append(res,points)
    return res.reshape((amount,dimension))

def get_random_points(data,amount): #Get a given amount of point randomly from given data
    which = []
    points = []
    for j in range(amount):
        index = random.randint(0,len(data)-1)
        while(index in which):
            index = random.randint(0,len(data)-1)

        which.append(index)
    for el in which:
        points.append(data[el])
    return points
    
def euclidean_distance(x,y): #Distance between points in the same dimension
    return np.linalg.norm(x-y)

def get_plot_parameters(arr): #Get mean and standard devation of passed results for plotting.
    return np.mean(arr), np.std(arr)
```

```python
#Functions to get an angle between two given vectors
def angle_between(v1, v2):
    u1 = v1 / np.linalg.norm(v1)
    u2 = v2 / np.linalg.norm(v2)
    return np.arccos(np.clip(np.dot(u1, u2),-1.0,1.0))
```

     One set of points -> 10 tries of getting angle between some chosen vectors
     Ten tries for one chosen dimension.

# Tasks from pdf

```python
#All functions return an array with results for given task. The amount of results is dependent on value of retries parameter
def first_task(data,retries):
    res = np.array([])
    for i in range(retries):
        points = get_random_points(data,4)  
        v1 = points[1] - points[0]
        v2 = points[3] - points[2]
        res = np.append(res,angle_between(v1,v2))
    return res

def second_task(data,retries):
    res = np.array([])
    for i in range(retries):
        inside = 0.0
        points = get_random_points(data,2)
        radius = euclidean_distance(points[1],points[0])
        for el in data:
            if euclidean_distance(points[0],el) < radius and (el != points[0]).all():
                inside+=1.0
        
        res = np.append(res,(inside/len(data)*1.0)*100.0)
    return res

def third_task(data,retries):
    res = np.array([])
    for i in range(retries):
        points = get_random_points(data,3)
        firstdistance = euclidean_distance(points[1],points[0])
        seconddistance = euclidean_distance(points[2],points[0])
        
        res = np.append(res,((firstdistance+seconddistance/2)/abs(firstdistance-seconddistance))*100)
    return res
```

# Run testing

```python
def run_dimension_tests(side,dimension,amount,retries,algorithm):
    data = generate_points(side,dimension,amount)
    res = np.ndarray(shape = (retries,retries))
    for i in range(retries):
        res[i] = algorithm(data,retries)
        
    return np.reshape(res,retries*retries)

def run_test(side,startdimension,enddimension,startamount,endamount,retries):
    results = {}
    for i in range(startdimension,enddimension):
        for j in range(startamount,endamount, (endamount-startamount)//20):
            res = (get_plot_parameters(run_dimension_tests(side,i,j,retries,first_task)),
                  get_plot_parameters(run_dimension_tests(side,i,j,retries,second_task)),
                  get_plot_parameters(run_dimension_tests(side,i,j,retries,third_task)))
            results[(i,j)] = res
    return results
```

Format of result map -> (dimension,pointsamount) : {res for first, res for second, res for third} where res -> average from retries and standard deviance

# Functions for plotting values

```python
def set_plot_parameters(xaxis,yaxis,errbar,plot_args):
    plt.clf()
    plt.rcParams["figure.figsize"] = plot_args['fig_size']
    plt.errorbar(xaxis,yaxis,errbar,fmt=plot_args['fmt'],ecolor=plot_args['errcolor'],color=plot_args['color'],elinewidth = 1)
    plt.title(plot_args['title'])
    plt.xlabel(plot_args['xlabel'])
    plt.ylabel(plot_args['ylabel'])


def plot_dimensions_static_size(mapping,startdim,enddim,amount,algorithm,plot_args):
    xaxis = []
    yaxis = []
    errbar = []
    for i in range(startdim,enddim):
        data = mapping[(i,amount)][algorithm]
        xaxis.append(i)
        yaxis.append(data[0])
        errbar.append(data[1])
    
    set_plot_parameters(xaxis,yaxis,errbar,plot_args)
    
def plot_sizes_static_dimensions(mapping,dim,algorithm):
    xaxis = []
    yaxis = []
    errbar = []
    for key,value in mapping.items():
        if key[0] == dim:
            data = value[algorithm]
            xaxis.append(key[1])
            yaxis.append(data[0])
            errbar.append(data[1])
    
    set_plot_parameters(xaxis,yaxis,errbar,plot_args)
    
def plot_3D_values(mapping,algorithm,plot_args,bars = False):
    fig = plt.figure(dpi=100)
    ax = fig.add_subplot(111, projection='3d')
    xaxis = []
    yaxis = []
    zaxis = []
    errbar = []
    for key,value in mapping.items():
        data = value[algorithm]
        xaxis.append(key[1])
        yaxis.append(key[0])
        zaxis.append(data[0])
        errbar.append(data[1])
        
    ax.plot(xaxis,yaxis,zaxis,marker=plot_args['fmt'],linestyle="None",color=plot_args['color'])
    if bars:
        for i in range(len(errbar)):
            ax.plot([xaxis[i], xaxis[i]], [yaxis[i], yaxis[i]], [zaxis[i]+errbar[i], zaxis[i]-errbar[i]], marker="_")
        
    ax.set_title(plot_args['title'])
    ax.set_xlabel(plot_args['xlabel'])
    ax.set_ylabel(plot_args['ylabel'])
    ax.set_zlabel(plot_args['zlabel'])
    
def return_available_sizes(mapping):
    values = []
    print("Available sizes of points set to show:")
    for key,value in mapping.items():
        if key[1] not in values:
            values.append(key[1])
    return values
```

```python
results = run_test(1,2,100,100,3000,10) #Calculating results
```
